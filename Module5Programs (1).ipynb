```python
# Required Header
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

## Create a undirected graph UG and show it

```python
UG = nx.fast_gnp_random_graph(15,0.3)
nx.draw(UG,with_labels=True)
```

## Create a directed graph DG and show it

```python
DG = nx.fast_gnp_random_graph(15,0.2,directed=True)
nx.draw(DG,with_labels=True)
```

## Show the adjacency matrix of UG and edge list of UG

```python
A = nx.adjacency_matrix(UG)
matA = A.todense()
print(matA)
# Print the edge list
print ('Edge List')
for line in nx.generate_edgelist(UG, data=False):
    print(line)
```

## Show the adjacency matrix of DG and the edge (arc) list of DG

```python
B = nx.adjacency_matrix(DG)
matB = B.todense()
print(matB)
# Print the edge list
print ('Arc List')
for line in nx.generate_edgelist(DG, data=False):
    print(line)
```

## Find if the graph UG is connected and DG is strongly connected

```python
if (nx.is_connected(UG)):
    print ('The undirected graph is connected')
else:
    print (f'The undirected graph is not connected and has {nx.number_connected_components(UG)} connected components')
if (nx.is_strongly_connected(DG)):
    print ('The directed graph is stronlgy connected')
else:
    print (f'The directed graph is not stronlgy connected and has {nx.number_strongly_connected_components(DG)} strongly connected components')    
```

## Find the connected components of the undirected graph

```python
components = nx.connected_components(UG)
for component in components:
    print(f'{len(component)}: {component}')
```

## Find the strongly connected components of the directed graph

```python
strong_components = nx.strongly_connected_components(DG)
for strong_component in strong_components:
    print(f'{len(strong_component)}: {strong_component}')
```

### Weighted Random Graph

```python
import random
WG = nx.fast_gnp_random_graph(10,0.3)
edgelist = []
for (u, v) in WG.edges():
    rand = random.randint(1,10)
    edgelist.append ((u,v,rand))
    edgelist.append ((v,u,rand))
print (edgelist)
```

```python
G2 = nx.Graph()
G2.add_nodes_from(WG.nodes())
G2.nodes()
G2.add_weighted_edges_from(edgelist)
```

```python
plt.axis('off')
pos = nx.spring_layout(G2,k=40,iterations=10)
nx.draw_networkx_nodes(G2,pos,node_color='g',alpha = 0.8)  # draws nodes
nx.draw_networkx_edges(G2,pos,edge_color='b',alpha = 0.3)  # draws edges
nx.draw_networkx_edge_labels(G2,pos,edge_labels = nx.get_edge_attributes(G2,'weight')) # edge lables
nx.draw_networkx_labels(G2,pos) # node lables
```

```python
sp = nx.dijkstra_path(G2,source = 0, target = 1)
print(sp)
```

### Shortest Path: http://avinashu.com/tutorial/pythontutorialnew/NetworkXBasics.html

```python
G4 = nx.Graph()
list_edges = [(1,2,2.0) , (1,4,4.0) , (2,3,3.0) , (2,5,2.0) , (3,6,1.0) , (4,5,2.0) , (4,7,6.0) , (5,6,5.0) , (5,8,3.0) , (6,9,8.0) , (8,9,1.0) , (7,8,2.0)]
G4.add_weighted_edges_from(list_edges)
G4.nodes[1]['pos'] = (0,0)
G4.nodes[2]['pos'] = (0,2)
G4.nodes[3]['pos'] = (0,4)
G4.nodes[4]['pos'] = (2,0)
G4.nodes[5]['pos'] = (2,2)
G4.nodes[6]['pos'] = (2,4)
G4.nodes[7]['pos'] = (4,0)
G4.nodes[8]['pos'] = (4,2)
G4.nodes[9]['pos'] = (4,4)
T = nx.minimum_spanning_tree(G4)
print(T.edges())
```

```python
node_pos=nx.get_node_attributes(G4,'pos')
edge_weight=nx.get_edge_attributes(G4,'weight')
red_edges = T.edges()
node_col = ['white']
# If the edge is in the shortest path set it to red, else set it to white color
edge_col = ['black' if not edge in red_edges else 'red' for edge in G4.edges()]
# Draw the nodes
nx.draw_networkx(G4, node_pos,node_color= node_col, node_size=450)
# Draw the node labels
# nx.draw_networkx_labels(G4, node_pos,node_color= node_col)
# Draw the edges
nx.draw_networkx_edges(G4, node_pos,edge_color=edge_col)
# Draw the edge labels
nx.draw_networkx_edge_labels(G4, node_pos, edge_labels=edge_weight)
# Remove the axis
plt.axis('off')
# Show the plot
plt.show()
G4.clear()
```

### Max Flow Algorithm: http://avinashu.com/tutorial/pythontutorialnew/NetworkXBasics.html

```python
G5 =nx.DiGraph()
G5.add_edge(1,2, weight=2, capacity=4)
G5.add_edge(1,3, weight=2, capacity=2)
G5.add_edge(2,3, weight=1, capacity=2)
G5.add_edge(2,4, weight=3, capacity=3)
G5.add_edge(3,4, weight=1, capacity=5)
G5.nodes[1]['demand'] = -4
G5.nodes[4]['demand'] = 4
cost, flow = nx.capacity_scaling(G5)
print("The minimum cost is:", cost)
print("The optimal flow is:")
print(flow)
# The data structure of flow is not consistent with dictionary datastructure
# needed for printing the optimal solution
optimal_flow = {}
for i in G5.nodes():
    for j in flow[i].keys():
        optimal_flow[i,j] = flow[i][j]
```

```python
G5.nodes[1]['pos'] = (0,0)
G5.nodes[2]['pos'] = (3,3)
G5.nodes[3]['pos'] = (3,-3)
G5.nodes[4]['pos'] = (6,0)
node_pos = nx.get_node_attributes(G5,'pos')
arc_cost = nx.get_edge_attributes(G5,'weight')
arc_cap  = nx.get_edge_attributes(G5,'capacity')
```

```python
nx.draw_networkx(G5, node_pos,node_color='white' , node_size=450)
#nx.draw_networkx_labels(G5, node_pos,node_color='white')
#nx.draw_networkx_edges(G5, node_pos)
nx.draw_networkx_edge_labels(G5, node_pos,edge_labels=optimal_flow)
# Remove the axis
plt.axis('off')
# Show the plot
print("The edge weights correspond to optimal flow values:")
plt.show()
```
