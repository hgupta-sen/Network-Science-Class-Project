d) Take the graph UG1. Take an edge (x, y); Remove (x, y) from UG1; call the new graph UG11; Find the number of connected components for UG11; Repeat this process for all edges (x, y) in UG1; FIND THE EDGE THAT RESULTS IN THE MAXIMUM NUMBER OF CONNECTED COMPONENTS


```python
edges = list(UG1.edges())
print(edges)
for i in range (len(UG1.edges())): 
    new_edges = edges[:i] + edges[i+1:]
    UG11 = nx.Graph() 
    UG11.add_edges_from(new_edges)
    undirected_components = nx.connected_components(UG11)
    for undirected_component in undirected_components:
        print(f'{len(undirected_component)}: {undirected_component}')
    print("Number of connected components while removing edge ", edges[i], "is ", nx.number_connected_components(UG11))
    print("\n\n")
```

### Comment: 


From the result, its visiable that when I am removing (4,11) or (11,14) the number of connected components are getting 3. In rest of the cases, the number of connected components are 2.

b) Take the graph DG1. Take an ARC (x, y); Remove (x, y) from DG1; call the new graph DG11; Find the number of **strongly connected** components for DG11; Repeat this process for all edges (x, y) in DG1; FIND THE **ARC** THAT RESULTS IN THE MAXIMUM NUMBER OF **Stronlgy CONNECTED** COMPONENTS

```python
directed_edges = list(DG1.edges())
print("The edges of the directed graph DG1 is as follows: ", directed_edges)
for i in range (len(DG1.edges())): 
    directed_new_edges = directed_edges[:i] + directed_edges[i+1:]
    DG11 = nx.DiGraph() 
    DG11.add_edges_from(directed_new_edges)
    strong_components = nx.strongly_connected_components(DG11)
    for strong_component in strong_components:
        print(f'{len(strong_component)}: {strong_component}')
    print("Number of strongly connected components while removing edge ", directed_edges[i], "is ", nx.number_strongly_connected_components(DG11)) 
    print("\n\n")       
```

### Comment:

From the result, its visiable that when I am removing (9, 16) the number of connected components are getting 7. In rest of the cases, the number of connected components are either 5 or 6.
