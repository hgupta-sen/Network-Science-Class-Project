## Find the degree of each node in G1 and draw a histogram: nodes on the x-axis and degree on the y-axis

```python
# Find of the dgree of each nodes
G1.degree () 

#Creating blank list for keeping nodes and coresponding degrees 
node=[]
degree=[]


for n,d in G1.degree () :    
    
    #Priting the degree of each and every nodes
    print("Degree of Node ", n, "is ", d)
    
    #Flling the blank lists 
    node.append(n)
    degree.append(d)
    
#Drawing and showing a histogram: nodes on the x-axis and degree on the y-axis
plt.bar(node, degree,  color ='maroon', width = 0.9)
plt.xlabel("Nodes")
plt.ylabel("Degree")
plt.title("Histogram (G1 Graph): Nodes VS Degree ")
plt.show()
```

## Find the average degree of nodes in G1

```python
# for n is like a for-loop goes through each element in G.degree. d is the degree of each node in G.degree() list
degree_sequence = [d for n, d in G1.degree()]  # degree sequence

#Getting average degree of the graph by deviding the total degree of the graph by the number of nodes
average_degree_of_nodes = sum(degree_sequence)/len(degree_sequence)

print("Average degree of nodes in G1 is ", average_degree_of_nodes)
```

## Find the number of edges in the graph G1

## Construct a G2 from G1 by adding edges randomly to nodes in G1 whose degree is less than 5; For example if node 10 has a degree 3, we add two new edges (in random). Note that when you add an edge (10, 19), node 19 also gains a degree.

```python
#Copying G1 to G2
G2 = G1.copy()

#Creating a blank list for storing nodes whose degree is less than 5 
nodes_less5=[]
for nodeG11, degreeG11 in G1.degree():      
    if degreeG11 < 5: 
        nodes_less5.append(nodeG11)

#Adding edges randoming to the edges whose degree is less than 5 
for nodeG1, degreeG1 in G1.degree():      
    if degreeG1 < 5: 
        edges_add = 5 - degreeG1  
        for i in range(edges_add+1): 
            G2.add_edge(nodeG1, random.choice(list(nodes_less5)))

#Checking the number of edges which should be increased due to adding multiple edges
number_edge2 = len(G2.edges())

print("The number of edges in the graph G1 is ", len(list(G1.edges())))

print("\n\nThe number of edges in the graph G2 is ", number_edge2)

print("\n\nThe number of nodes in the graph G1 is ", len(list(G1.nodes())))

print("\n\nThe number of nodes in the graph G2 is ", len(list(G2.nodes())))

for nodeG2, degreeG2 in G2.degree():     
    
    if degreeG2 < 5: 
        
        print("Degree of", nodeG2, "is", degreeG2)

#Checking the minimum degree of G2 to ensure that I successfully increased the degree of all the nodes where degree is less than 5
print("\n\nThe minimum degree of the graph G2 is ", min([degreeG2 for nodeG2, degreeG2 in G2.degree()]))
```

## Construct a subgraph H2 from H1 by removing nodes whose degree is less than 2 times the average degree and nodes whose degree is greater than 2 times the avearge degree

```python
H2 = H1.copy()

print("Nodes of H2 before removing nodes: ", H2.nodes())
for nodeH1, degreeH1 in H1.degree () : 
    if degreeH1 < 2 * average_degree_of_nodes1: 
        H2.remove_node(nodeH1)
        
    elif degreeH1 > 2 * average_degree_of_nodes1: 
        H2.remove_node(nodeH1)
print("\n\nNodes of H2 before removing nodes: ", H2.nodes())    

print("\n\nThe Number of Edges in the Graph H1 is ", len(H1.edges()))

print("\n\nThe Number of Edges in the Graph H2 is ", len(H2.edges()))
```
